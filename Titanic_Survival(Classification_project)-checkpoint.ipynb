# missing data treatment & Data Preparation

```python

Q=pd.DataFrame(X.isna().sum(),columns=["CT"])
W=Q[Q.CT> 0].index
for i in W:
    if X[i].dtypes=="object":
        R=X[i].mode()[0]
        X[i]=X[i].fillna(R)
    else:
        R=X[i].mean()
        X[i]=X[i].fillna(R)
```

```python
X.isna().sum()
```

```python
con=[]
cat=[]
for i in X.columns:
    if X[i].dtypes=="object":
        cat.append(i)
    else:
        con.append(i)
X1=pd.get_dummies(X[cat])
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X2=pd.DataFrame(ss.fit_transform(X[con]),columns=con)
Xnew=X2.join(X1)
```

```python
Xnew
```

# Devide data into training and testing set

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(Xnew,Y,test_size=0.2,random_state=21)
```

# Making Models
1) Logistic Regression
2)Decision Tree Classifier
3)AdaBoostClassifier

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
model=lr.fit(Xtrain,Ytrain)
tr_pred=model.predict(Xtrain)
from sklearn.metrics import accuracy_score,confusion_matrix
tr_acc=accuracy_score(Ytrain,tr_pred)
```

```python
tr_acc
```

```python
confusion_matrix(Ytrain,tr_pred)
```

# Making Predictions

```python
Ytrain[["predicted"]]=tr_pred
```

```python
Ytrain
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC=DecisionTreeClassifier(max_depth=4,random_state=21)
model=DTC.fit(Xtrain,Ytrain)
tr_pred=model.predict(Xtrain)
```

```python
accuracy_score(Ytrain,tr_pred)
```

```python
from sklearn.ensemble import AdaBoostClassifier
adc=AdaBoostClassifier(DecisionTreeClassifier(max_depth=4),random_state=21)
model=adc.fit(Xtrain,Ytrain)
tr_pred=model.predict(Xtrain)
```

```python
Ytrain
```

```python
accuracy_score(Ytrain,tr_pred)
```

```python
Ytrain[["predicted"]]=tr_pred
```
